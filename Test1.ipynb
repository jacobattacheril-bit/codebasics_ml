### Content by Country

```python
# The 'country' column can contain multiple countries, so we need to split them
# First, handle missing values by filling with an empty string or 'Unknown'
df_kaggle_countries = df_kaggle.copy()
df_kaggle_countries['country'].fillna('Unknown', inplace=True)

# Split all country strings into individual countries and flatten the list
all_countries = []
for countries_str in df_kaggle_countries['country']:
    # Ensure to remove spaces and split, and filter out empty strings
    all_countries.extend([c.strip() for c in countries_str.split(',') if c.strip()])

# Count occurrences of each country
country_counts = pd.Series(all_countries).value_counts()

# Get the top 10 countries
top_10_countries = country_counts.nlargest(10)

# Visualize the top 10 countries
plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_countries.index, y=top_10_countries.values, palette='plasma')
plt.title('Top 10 Countries Producing Content on Netflix')
plt.xlabel('Country')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### Most Popular Genres / Listed In Categories

```python
# The 'listed_in' column contains multiple genres, so we need to split them
all_genres = []
for genres_str in df_kaggle['listed_in']:
    all_genres.extend([g.strip() for g in genres_str.split(',') if g.strip()])

# Count occurrences of each genre
genre_counts = pd.Series(all_genres).value_counts()

# Get the top 10 genres
top_10_genres = genre_counts.nlargest(10)

# Visualize the top 10 genres
plt.figure(figsize=(12, 7))
sns.barplot(x=top_10_genres.index, y=top_10_genres.values, palette='cubehelix')
plt.title('Top 10 Genres on Netflix')
plt.xlabel('Genre')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45, ha='right')
plt.show()
```

### ML-Based Insight: Content-Based Recommendation System

To build a content-based recommendation system, we need to first create a feature set that describes each title. We can start by combining relevant text-based columns like `listed_in` and `description` to create a content profile for each show.

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import linear_kernel

# Fill any potential NaN values in 'description' or 'listed_in' with empty strings
df_kaggle['description'] = df_kaggle['description'].fillna('')
df_kaggle['listed_in'] = df_kaggle['listed_in'].fillna('')

# Create a 'content_features' column by combining relevant textual data
# We'll use 'listed_in' (genres) and 'description' for now
df_kaggle['content_features'] = df_kaggle['listed_in'] + ' ' + df_kaggle['description']

# Initialize TF-IDF Vectorizer
# TF-IDF measures how important a word is to a document in a collection of documents
tfidf_vectorizer = TfidfVectorizer(stop_words='english')

# Construct the TF-IDF matrix
tfidf_matrix = tfidf_vectorizer.fit_transform(df_kaggle['content_features'])

print("TF-IDF matrix shape:", tfidf_matrix.shape)
```

Now that we have the TF-IDF matrix, we can compute the cosine similarity between titles. Cosine similarity measures the cosine of the angle between two non-zero vectors. It is often used to measure document similarity in text analysis.

```python
# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

print("Cosine similarity matrix shape:", cosine_sim.shape)

# Create a series with title and index to easily retrieve indices from titles
title_to_index = pd.Series(df_kaggle.index, index=df_kaggle['title']).drop_duplicates()

def get_recommendations(title, cosine_sim_matrix=cosine_sim, df=df_kaggle, indices=title_to_index):
    # Get the index of the movie that matches the title
    idx = indices[title]

    # Get the pairwise similarity scores of all movies with that movie
    sim_scores = list(enumerate(cosine_sim_matrix[idx]))

    # Sort the movies based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar movies (excluding itself)
    sim_scores = sim_scores[1:11]

    # Get the movie indices
    movie_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar movies
    return df['title'].iloc[movie_indices]

# Example: Get recommendations for 'Kota Factory'
print("\nRecommendations for 'Kota Factory':")
display(get_recommendations('Kota Factory'))
```
